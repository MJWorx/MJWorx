## Income vs Spending Score

```python
sns.scatterplot(x='Income', y='Spending_Score', data=df, hue='Age', palette='viridis')
plt.title('Income vs Spending Score')
plt.show()
```

## Membership Years vs Last Purchase Amount

```python
sns.set_style('whitegrid')
sns.scatterplot(x='Membership_Years', y='Last_Purchase_Amount', data=df, hue='Purchase_Frequency', palette='Set3_r')
plt.title('Membership Years vs Last Purchase Amount')
plt.show()
```

## Last Purchase Amount by Membership Years

```python
sns.set(rc = {'figure.figsize': (20, 10)})
sns.boxplot(x='Membership_Years', y='Last_Purchase_Amount', data=df, palette='terrain_r')
plt.title('Last Purchase Amount By Membership Years', fontsize=25, fontweight='bold')
plt.xlabel('Membership_Years', fontsize=20)
plt.ylabel('Last_Purchase_Amount', fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```

```python
plt.figure(figsize = (15, 10))
numeric_cols = ['Age', 'Income', 'Spending_Score', 'Membership_Years',
       'Purchase_Frequency', 'Last_Purchase_Amount']

for i, col in enumerate(numeric_cols):
    plt.subplot(3, 3, i + 1)
    sns.histplot(data = [col], kde=True, color='crimson')
    plt.title(f'Distribution of {col}')
plt.tight_layout()
plt.show()
```

## Correlation Heatmap

```python
df.corr()
```

```python
plt.figure(figsize = (20, 10))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap', fontsize=25, fontweight='bold')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
