# Data Cleaning In Pandas

```python
import pandas as pd
```

```python
df = pd.read_excel('Customer Call List.xlsx')
```

```python
df
```

```python
df = df.drop_duplicates()
df
```

```python
df = df.drop(columns = "Not_Useful_Column")
df
```

```python
df['Last_Name'].str.strip()
```

```python
df['Last_Name'].str.lstrip("...")
```

```python
df['Last_Name'].str.lstrip("/")
```

```python
# to make permanant change to df
df['Last_Name'] = df['Last_Name'].str.lstrip("...")
df
```

```python
# make permanant changes to the df
df['Last_Name'] = df['Last_Name'].str.lstrip("/")
df
```

```python
# to make permanant change to df
df['Last_Name'] = df['Last_Name'].str.rstrip("_")
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('[^a-zA-Z0-9]', '',regex=True)
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
df
```

```python
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
df
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('nan--', '')
df
```

```python
df['Phone_Number'].str.replace('Na--', '')
```

```python
df['Phone_Number'] = df['Phone_Number'].str.replace('Na--', '')
df
```

```python
df['Address'].str.split(',', n=1, expand=True)
```

```python
df['Address'].str.split(',', n=2, expand=True)
```

```python
df[['Street_Address', 'State', 'Zip_Code']] = df['Address'].str.split(',', n=2, expand=True)
df
```

```python
df['Paying Customer'].str.replace('Yes', 'Y')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('Yes', 'Y')
df
```

```python
df['Paying Customer'].str.replace('No', 'N')
```

```python
df['Paying Customer'] = df['Paying Customer'].str.replace('No', 'N')
df
```

```python
df['Do_Not_Contact'].str.replace('Yes', 'Y')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('Yes', 'Y')
df
```

```python
df['Do_Not_Contact'].str.replace('No', 'N')
```

```python
df['Do_Not_Contact'] = df['Do_Not_Contact'].str.replace('No', 'N')
df
```

```python
df = df.replace('N/a', '')
df
```

```python
df = df.replace('NaN', '')
df
```

```python
df = df.fillna('')
df
```

```python
# getting the 'Y' out of the df from column 'Do_Not_Contact'
for x in df.index:
    if df.loc[x, 'Do_Not_Contact'] == 'Y':
        df.drop(x, inplace=True)

df
```

```python
# getting the '' out of the df from column 'Phone_Number'

for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace=True)

df
```

```python
# dropping our original index from the df

df = df.reset_index(drop=True)
df
```
