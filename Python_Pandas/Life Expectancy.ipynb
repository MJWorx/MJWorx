## Visualization

### Life Expectancy Distribution by Region

#### Boxplot

```python
plt.figure(figsize=(15, 12))
sns.boxplot(x='Region', y='Life_expectancy', data=df, palette='Paired')
plt.title('Life Expectancy Distribution By Region')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

### Countries Low Mean Life Expectancy

#### Barplot

```python
df_des = df['Life_expectancy'].describe()
df_des
```

```python
life_exp_less50 = df[df['Life_expectancy'] < df_des['mean']]
life_exp_less50
```

```python
plt.figure(figsize=(15, 6))
plt.bar(life_exp_less50['Country'], life_exp_less50['Life_expectancy'], color='turquoise')
plt.title('Countries With Life Expectancy Below Mean')
plt.xlabel('Countries')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.show()
```

#### Map Of Countries Below Mean

```python
fig = px.choropleth(life_exp_less50, locationmode='country names', color='Life_expectancy', hover_name='Country', hover_data= {'Region': True, 'Life_expectancy': True},
                   color_continuous_scale='viridis', title='Average Life Expectancy By Country')
fig.show()
```

#### Top 10 Countries By High Life Expectancy

```python
top_10_life_exp = df[['Country', 'Life_expectancy']].sort_values(by='Life_expectancy', ascending=False).head(10)
top_10_life_exp
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Life_expectancy', y='Country', data= top_10_life_exp, palette='magma', orient='h')
plt.title('Top 10 Countries By High Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Country')
plt.xticks(rotation=90)
plt.show()
```

#### Key Health Indicators

```python
key_vars = ['Life_expectancy', 'GDP_per_capita', 'Alcohol_consumption', 'BMI', 'Incidents_HIV']
sns.pairplot(df[key_vars], diag_kind='kde')
plt.suptitle('Key Health Indicators', y=1.02)
plt.show()
```

#### Distribution Of Life Expectancy By Fequency

```python
plt.figure(figsize=(10, 5))
sns.histplot(df['Life_expectancy'], palette= sns.set_palette('viridis'), kde=True)
plt.title('Distribution Of Life Expectancy')
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.show()
```

#### GDP per Capita vs Life Expectancy by Region

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='GDP_per_capita', y='Life_expectancy', hue='Region', palette='terrain')
plt.xlabel('GDP_per_capita')
plt.ylabel('Life Expectancy')
plt.xticks(rotation=90)
plt.title('GDP Per Capita vs Life Expectancy By Region')
plt.legend(title='Region')
plt.show()
```

### Correlation Heatmap

```python
num_df = df.select_dtypes(include=[np.number])
num_df
```

```python
plt.figure(figsize=(18, 8))
sns.heatmap(num_df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix Of Life Expectancy')
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.show()
```
